# Удаление столбцов и перенос в другую папку


```python
import os
import pandas as pd

# Указываем пути
input_folder = '3ОФ34 - 120mm'
output_folder = 'koord_new'

# Создаем папку для сохранения новых файлов, если она еще не существует
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Удаляем ненужные столбцы из каждого файла
for filename in os.listdir(input_folder):
    # Проверяем, что файл имеет нужный формат
    if filename.startswith("Траектория 3ОФ34 угол") and filename.endswith(".txt"):
        # Полный путь к файлу
        filepath = os.path.join(input_folder, filename)
        
        # Читаем файл, пропуская первую строку с заголовками
        df = pd.read_csv(filepath, delim_whitespace=True)
        
        # Оставляем только нужные столбцы
        df = df[['t(c)', 'x(м)', 'y(м)', 'z(м)']]
        
        # Сохраняем файл в новую папку с тем же именем без заголовков столбцов
        output_filepath = os.path.join(output_folder, filename)
        df.to_csv(output_filepath, sep=' ', index=False, header=False)
        
print("Файлы успешно обработаны и сохранены в папку 'koord_new'")
```

# Замена запятых на точки

```python
import os

# Путь к папке с файлами
output_folder = 'koord_new'

# Обработка каждого файла в папке
for filename in os.listdir(output_folder):
    # Проверка, что это текстовый файл
    if filename.endswith(".txt"):
        filepath = os.path.join(output_folder, filename)
        
        # Чтение содержимого файла
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # Замена запятых на точки
        content = content.replace(',', '.')
        
        # Запись измененного содержимого обратно в файл
        with open(filepath, 'w', encoding='utf-8') as file:
            file.write(content)

print("Запятые успешно заменены на точки во всех файлах.")
```

# Нормировка

```python
import os
import pandas as pd

# Путь к папке с обработанными файлами
output_folder = 'koord_new'
# Имя файла для записи максимумов
max_values_file = 'max_values.txt'

# Открываем файл для записи максимумов
with open(max_values_file, 'w', encoding='utf-8') as max_file:
    # Обработка каждого файла в папке koord_new
    for filename in os.listdir(output_folder):
        # Проверка, что это текстовый файл
        if filename.endswith(".txt"):
            filepath = os.path.join(output_folder, filename)
            
            # Чтение и замена запятых на точки
            with open(filepath, 'r', encoding='utf-8') as file:
                content = file.read().replace(',', '.')
            
            # Перезапись файла с точками
            with open(filepath, 'w', encoding='utf-8') as file:
                file.write(content)
            
            # Чтение данных после замены
            df = pd.read_csv(filepath, delim_whitespace=True)
            
            # Преобразование всех значений в числовые (если возможно)
            df = df.apply(pd.to_numeric, errors='coerce')
            
            # Исправление первой строки (если нужно)
            first_row = df.iloc[0].copy()
            first_row[:] = [0.0000, 0.0000, 1.0000, 0.0000]  # Устанавливаем нужные значения
            
            df.iloc[0] = first_row  # Перезаписываем первую строку

            # Нахождение максимума по каждому столбцу, исключая NaN
            max_values = df.max(skipna=True)
            
            # Нормализация данных
            df_normalized = df / max_values
            
            # Перезапись файла с нормализованными данными
            df_normalized.to_csv(filepath, sep=' ', index=False, header=False)  # Без заголовков
            
            # Запись максимумов в файл max_values.txt
            max_values_str = ' '.join(map(str, max_values.values))  # Преобразуем максимумы в строку
            max_file.write(f"{filename}: {max_values_str}\n")  # Запись строки максимумов в файл

print("Файлы нормализованы, и максимальные значения сохранены в 'max_values.txt'.")
```
